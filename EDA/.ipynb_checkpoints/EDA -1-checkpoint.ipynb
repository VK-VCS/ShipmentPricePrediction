```python
l_weight=[]
l_Freight=[]
for i in range(0,df.shape[0]):
    if not(isFloat(df.iloc[i]['Weight (Kilograms)'])):
        l_weight.append(i)
    if not(isFloat(df.iloc[i]['Freight Cost (USD)'])):
        l_Freight.append(i)
print(len(l_weight))
print(len(l_Freight))
```

```python
for i in range(0,df.shape[0]):#loop through each row
    if not(isFloat(df.iloc[i]['Weight (Kilograms)'])): #check for non numeric data
        if re.search("^See", df.iloc[i]['Weight (Kilograms)']):# check for rows where refer to other row was mentioned
            idnum=re.split("ID#:*(\d+)*[)]", df.iloc[i]['Weight (Kilograms)'])[1] #get the reference row
            val=data[data['ID']==int(idnum)]['Weight (Kilograms)'] # get the weight value from reference row
            df.loc[i,'Weight (Kilograms)']=val.values #assign weight back to actual row

    if not(isFloat(df.iloc[i]['Freight Cost (USD)'])):
        if re.search("^See", df.iloc[i]['Freight Cost (USD)']):
            idnum=re.split("ID#:*(\d+)*[)]", df.iloc[i]['Freight Cost (USD)'])[1]
            val=data[data['ID']==int(idnum)]['Freight Cost (USD)']
            df.loc[i,'Freight Cost (USD)']=val.values
```

```python
l_weight=[]
l_Freight=[]
for i in range(0,df.shape[0]):
    if not(isFloat(df.iloc[i]['Weight (Kilograms)'])):
        l_weight.append(i)
    if not(isFloat(df.iloc[i]['Freight Cost (USD)'])):
        l_Freight.append(i)
print(len(l_weight))
print(len(l_Freight))
```

```python
import numpy as np
for i in range(0,df.shape[0]):
    if not(isFloat(df.iloc[i]['Weight (Kilograms)'])):
        df.loc[i,'Weight (Kilograms)']=np.nan
    if not(isFloat(df.iloc[i]['Freight Cost (USD)'])):
        df.loc[i,'Freight Cost (USD)']=np.nan
```

```python
for i in df.select_dtypes(include='object'): #checking for unique values in object columns
    print(f'{i}:{len(df[i].unique())}')
```
