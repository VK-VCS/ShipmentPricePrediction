###  Impute missing and zero values 

```python
import numpy as np
nonzerodata=data
zerocols=['Line Item Value','Pack Price','Unit Price','Weight (Kilograms)','Line Item Insurance (USD)']
nonzerodata[zerocols]=nonzerodata[zerocols].replace({0.0:np.nan})
```

```python
#Drop column - since all these 3 columns has < 20% missing values, we don't need to drop these columns
#guess value and forward back ward imputation - not advisable as this data doesn't follow any sorting trend
#random value imputation - as data range is large, random values imputation will not be effective
#filling with central tendecy measures - not advisable as data range is large
#create new feature for each column to indicate null or not null - increases 3 more columns,not adivsable
#replacing with 2,3rd Standard deviation values - relation between other columns will be missing
#use model for prdicting missing values - KNN Imputer
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
non_nulldata=pd.DataFrame(imputer.fit_transform(nonzerodata,nonzerodata.columns),columns=nonzerodata.columns)
non_nulldata.head()
```

```python
profile = pp.ProfileReport(non_nulldata)
#profile.to_file("F:\\ML\\ML Ineuron\\Project\\ShipmentPricingPrediction\\EDA\\Profiled_Data2.html")
```

```python
#method2: applying log
logData=non_nulldata
cols=['Unit of Measure (Per Pack)','Line Item Quantity','Line Item Value','Pack Price','Unit Price','Weight (Kilograms)','Freight Cost (USD)','Line Item Insurance (USD)']
logData[cols]=np.log(logData[cols])
#profile = pp.ProfileReport(logData)
#profile.to_file("F:\\ML\\ML Ineuron\\Project\\ShipmentPricingPrediction\\EDA\\Profiled_Data4.html")
logData.head()
#conclusion: Unit of Measure (Per Pack),Pack Price,Unit Price - still not normally distributed, so need to remove outliers on these columns
```

```python
#method3: removing outliers 
cols=['Unit of Measure (Per Pack)','Line Item Quantity','Line Item Value','Pack Price','Unit Price','Weight (Kilograms)','Freight Cost (USD)','Line Item Insurance (USD)']
sns.set()
%matplotlib inline
fig, ax = plt.subplots(figsize=(20,10))
sns.boxplot(data=logData[cols], width=0.5,ax=ax)
```

```python
q = logData['Unit Price'].quantile(0.99)
data_cleaned = logData[logData['Unit Price']<q]
```

```python
sns.distplot(data_cleaned['Unit Price'])
```

```python
q = data_cleaned['Pack Price'].quantile(0.9)

data_cleaned=data_cleaned[data_cleaned['Pack Price']<q]
data_cleaned=data_cleaned[data_cleaned['Pack Price']>data_cleaned['Pack Price'].quantile(0.01)]
```

```python
sns.distplot(data_cleaned['Pack Price'])
```

```python
q=data_cleaned['Unit of Measure (Per Pack)'].quantile(0.99)
data_cleaned=data_cleaned[data_cleaned['Unit of Measure (Per Pack)']<q]
data_cleaned=data_cleaned[data_cleaned['Unit of Measure (Per Pack)']>data_cleaned['Unit of Measure (Per Pack)'].quantile(0.05)]
```

```python
sns.distplot(data_cleaned['Unit of Measure (Per Pack)'])
```

```python
data_cleaned=data_cleaned[data_cleaned['Line Item Quantity']>data_cleaned['Line Item Quantity'].quantile(0.03)]
data_cleaned=data_cleaned[data_cleaned['Line Item Value']>data_cleaned['Line Item Value'].quantile(0.0001)]
```

```python
data_cleaned=data_cleaned[data_cleaned['Weight (Kilograms)']>data_cleaned['Weight (Kilograms)'].quantile(0.005)]
data_cleaned=data_cleaned[data_cleaned['Weight (Kilograms)']<data_cleaned['Weight (Kilograms)'].quantile(0.999)]
data_cleaned=data_cleaned[data_cleaned['Line Item Insurance (USD)']>data_cleaned['Line Item Insurance (USD)'].quantile(0.005)]
```
